```python
b= np.array([10,0])
A=np.array([[10, 0], [0, 1]])
x=np.array([2.0,10])

for i in range(0,10):
    p_k = -np.linalg.inv(A).dot(A.dot(x)-b)
     
    alpha=(p_k.transpose().dot(p_k))/(p_k.transpose().dot(A.dot(p_k)))
    x=x+p_k
    print(x)
```

# Question 4

```python
arr=np.zeros(300)
for j in range(300):

    x=20
    k=j
    def grad(x):
        return x**30/(30*x**29)
    for i in range(k):
        p_k=grad(x)
        x=x-p_k
    arr[k]=x


    
```

```python
yaxis=np.arange(0,300)
xaxis=arr

fig = plt.figure(figsize=(8,5))
ax = plt.axes()


ax.scatter(xaxis,yaxis,marker='X');

ax.set(xlabel='Distance', ylabel='Iterations',
       title='Distance vs iterations');
```
